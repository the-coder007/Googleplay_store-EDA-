***we can convert the size column to numerical columns because it contain numerical value with some str extension***

```python
df_copy["Size"]
```

```python
#for checking unique value
df_copy["Size"].unique()
```

```python
#for converting size in MB to KB
```

```python
df_copy["Size"]=df_copy["Size"].str.replace('M','000')
```

```python
df_copy["Size"]=df_copy["Size"].str.replace('k',' ')
```

```python
df_copy["Size"]
```

```python
df_copy["Size"].unique()
#sucessfull change
#but there is one row with str, so we will replace it with missibg value
```

```python
#replace 
df_copy["Size"]=df_copy["Size"].str.replace('Varies with device',str(np.nan))
```

```python
df_copy["Size"].unique()
```

```python
#for converting the datatype of size row
```

```python
df_copy["Size"]=df_copy["Size"].astype("float")
```

```python
df_copy["Size"].dtype
```

```python
df_copy["Size"].head()
```

***we have some problem in the data after removing M from the size of data we are getting some data which earlier 1.2M but after Removing M it become 1.2 only so for this we have to do handle it***

```python
for i in df_copy["Size"]:
    if i<10:
        df_copy["Size"]=df_copy["Size"].replace(i,i*1000)
```

```python
df_copy["Size"].head()
```

```python
df_copy["Size"].unique()
```

```python
#we can short the size value or scale our data
```

```python
df_copy["Size"]=df_copy["Size"]/1000
```

```python
#or we can write the function also to clean these data columns (installs and price)
```

```python
char_to_remove=["+",",","$"]
col_to_clean=["Installs","Price"]
for items in char_to_remove:
    for col in col_to_clean:
        df_copy[col]=df_copy[col].str.replace(items,"")
```
